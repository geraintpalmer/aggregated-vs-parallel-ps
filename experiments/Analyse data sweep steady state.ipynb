```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import traces
```

```python
def plot_ma(R, rho, ax, c, window):
    series = traces.TimeSeries.from_csv(
        filename=f'sweep/simulation/R={R}-rho={rho}/sim0_recs.pq',
        time_column=4,
        time_transform=float,
        value_column=12,
        value_transform=int
    )
    series_mv = series.moving_average(window, pandas=True)
    ax.plot(series_mv, c=c, label=f"œÅ = {rho}")
```

```python
def plot_ts(R, window):
    fig, ax = plt.subplots(1)
    cmap = plt.get_cmap('viridis')
    for rho in [0.05*i for i in range(1, 20)]:
        plot_ma(R=R, rho=round(rho, 2), ax=ax, c=cmap(rho), window=1000)
    ax.set_title(f"R = {R}", fontsize=16)
    ax.legend(fontsize=8, bbox_to_anchor=(1, 1))
    return fig
```
