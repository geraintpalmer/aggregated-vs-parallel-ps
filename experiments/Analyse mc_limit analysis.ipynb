```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.pylab as pl
import numpy as np
plt.style.use('seaborn-whitegrid')
```

```python
cdfs = pd.read_csv('data/effect_mclimit_cdfs_R3.csv', index_col=0)
```

```python
wass = list(pd.read_csv('data/effect_mclimit_wasserstein_R3.csv'))
wass = [float(x) for x in wass]
```

```python
runtimes = list(pd.read_csv('data/effect_mclimit_runtime_R3.csv'))
runtimes = [float(x) for x in runtimes]
```

```python
mclims = range(2, 14)
fig, axarr = plt.subplots(2, 1, figsize=(12, 8))
axarr[0].plot(mclims, wass, c='black')
axarr[0].scatter(mclims, wass, c='black')
axarr[0].set_ylabel('Wassertein Distance', fontsize=16)
axarr[1].plot(mclims, runtimes, c='black')
axarr[1].scatter(mclims, runtimes, c='black')
axarr[1].set_ylabel('Runtimes (s)', fontsize=16)
axarr[0].set_xlabel('Markov Chain Limit', fontsize=16)
axarr[1].set_xlabel('Markov Chain Limit', fontsize=16)
axarr[0].set_xticks(mclims)
axarr[1].set_xticks(mclims)
axarr[0].set_title("$R=3$, $\lambda=14.5$, $\mu=6.2$, infty=130", fontsize=20)
plt.show()
```

```python
fig, ax = plt.subplots(1, figsize=(12, 6))
colors = pl.cm.viridis_r(np.linspace(0,1,len(mclims)))
ax.set_title("$R=3$, $\lambda=14.5$, $\mu=6.2$, infty=130", fontsize=20)
for i, mclim in enumerate(mclims):
    ax.plot(cdfs[str(mclim)], c=colors[i], label=f"limit={i+2}")
s = ax.plot(cdfs['Simulation'], c='black')
legendsim = plt.legend(s, ["Simulation"], fontsize=18, loc=8)
ax.legend(fontsize=12, loc=4)
plt.gca().add_artist(legendsim)
plt.show()
```
